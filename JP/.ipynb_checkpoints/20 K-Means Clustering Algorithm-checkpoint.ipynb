### Python Implementation of KMeans

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import numpy as np
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
```

```python
X[0:2]
```

```python
X.shape
```

```python
y[0:3]
```

```python
# visualize the dataset
```

```python
plt.scatter(X[:, 0], X[:, 1], s=20) # s is mark size
plt.show()
```

```python
model = KMeans(n_clusters=4)
model.fit(X)
y_pred = model.predict(X)
```

```python
# visualizing the cluster’s centers picked by k-means Python estimator
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=20, cmap='summer')
# s : float or array-like, shape (n, ), optional, The marker size in points**2.
# c : array-like or list of colors or color, optional, The marker colors. Possible values:
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=100, alpha=0.9);
plt.show()
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
type(digits)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(digits.data)
```

```python
df.head(2)
```

```python
df.shape
```

```python
# model fitting

kmeans=KMeans(n_clusters=10, random_state=0)
```

```python
clusters=kmeans.fit_predict(digits.data)
```

```python
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
from scipy.stats import mode
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```
