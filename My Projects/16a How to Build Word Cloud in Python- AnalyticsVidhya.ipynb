### How To Create Word Cloud in Python

```python
# 1.Import Necessary Libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 2.Selecting the Dataset
df = pd.read_csv("android-games.csv")
```

```python
# 3. Selecting the Text and Amount of Text for Word Cloud
```

```python
df.isna().sum()
```

```python
# 4. Adding Text to a Variable
text=" ".join(x.split()[1] for x in df.category)
```

```python
# 5. Creating the Word Cloud
word_cloud=WordCloud(max_words=100,collocations=False, background_color='black').generate(text)
```

```python
# 6. Plotting the Word Cloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
