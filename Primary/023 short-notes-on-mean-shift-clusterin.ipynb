### Mean Shift Algorithm : Definition and Features
- It's a centroid-based algorithm that tries to discover clusters in a datasets based on provided **bandwidth**
- It is also known as the mode seeking algorithm

> It assigns the datapoints to the clusters iteratively by shifting points towards the highest density of datapoints i.e. towards mode or centroid of the clusters.

>Unlike K-means clustering, it does not make any assumptions; hence it is a **non-parametric algorithm**.

>And this is the only diffrence between **Kmeans Clustering** and **Mean Shift clustering** that it doesn't need to specify numbers of cluster in advance.

#### How does Mean-Shift Algorithm work?
1. First, start with the data points assigned to a cluster usually
2. Then centroids are calculated according to density points
3. Then, points are shifted towards the mean or mode of the region where density of points is higher, and then location of new centroids is calculated.
4. Now this process of shifting points to a region of high density near them and calculation of **new centroids** after assigning **new points** is repeated recursively.
5. When it is not possible to relocate the centroids i.e. all centroids have reached to thier optimum position of higher density around them, the process is stopped.

#### Working Explanation 2:

Given a set of data points, the algorithm iteratively assigns each data point towards the closest cluster centroid and direction to the closest cluster centroid is determined by where most of the points nearby are at. 

So each iteration each data point will move closer to where the most points are at, which is or will lead to the cluster center. When the algorithm stops, each point is assigned to a cluster.

### Advantages
- **Simple to work** with as we don't need to calculate no of clusters in advance
- We just need to provide a **bandwidth** parameter, using this parameter the algorithm automatically discovers new clusters.
- It is **more efficeint** than Kmeans, as it can discover clusters that other model can't
- It can find clusters of **unusual shapes**
- It doesn't make any assumption prior to model fitting as like in K-means, so its **Non parametric algorithm**
- There is no issue of local minima as like in K-means.
- Can **handle outliers**,robust to outliers

#### Disadvantages
- Not fit to use with datasets having high dimensions
- May not find optimum clusters in high dimensional dataset
- We do not have more parameters to optimize
- It is computationally expensive of order O(nÂ²).

### Example 1:

```python
%matplotlib inline
import numpy as np
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
from sklearn.datasets import make_blobs
centers_points = [[2,2,2],[5,4,5],[6,8,10]]
X, y = make_blobs(n_samples = 700, centers = centers_points, cluster_std = 0.5)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
model = MeanShift()
model.fit(X)
labels = model.labels_

cluster_centers = model.cluster_centers_
print(cluster_centers)
no_clusters = len(np.unique(labels))
print("Estimated clusters:", no_clusters)

colors = 10*['r.','g.','b.','c.','k.','y.','m.']

for i in range(len(X)):
        plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 3)
        plt.scatter(cluster_centers[:,0],cluster_centers[:,1],
                    marker=".",color='k', s=20, linewidths = 5, zorder=10)
plt.show()
```
