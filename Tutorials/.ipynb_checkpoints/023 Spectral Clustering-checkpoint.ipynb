### Spectral Clustering

- uses the eigenvalues of a similarity matrix to reduce the dimensionality of the data and then applies clustering techniques in the reduced space
- useful for clustering data points that do not form compact clusters in the original space, such as data with non-linear or complex structures

All other algorithms used compactness(distance) between the data points as a characteristic to cluster the data points.

In Spectral Clustering, Using connectivity we can cluster two data points into the same clusters even if the distance between the two data points is larger.

It creates clusters based on the density of data points and their reachability within a specified distance parameter.

There are two major approaches in clustering. They are:

1. Compactness
2. Connectivity

In compactness, the points are closer to each other, compact toward centre, Distance is used as a measure to compute closeness. 

In connectivity, the points in a cluster are either immediately next to each other (epsilon distance) or connected.

Even if the distance is less, they might not be put in the same cluster.

How to do Spectral Clustering?
The three major steps:
- constructing a similarity graph,
- projecting data onto a lower-dimensional space
- clustering the data


Given a set of points S in a higher-dimensional space, it can be elaborated as follows:

1. Form a distance matrix
2. Transform the distance matrix into an affinity matrix A
3. Compute the degree matrix D and the Laplacian matrix L = D â€“ A.
4. Find the eigenvalues and eigenvectors of L.
5. With the eigenvectors of k largest eigenvalues computed from the previous step form a matrix.
6. Normalize the vectors.
7.Cluster the data points in k-dimensional space.

Pros and Cons of Spectral Clustering:
- In spectral, the clusters do not follow a fixed shape or pattern
- effective for data of different shapes and sizes
- computationally fast for sparse datasets of several thousand data points


Step by step working: 

1. Reachability Distance:
Calculation, This distance measures how far a point is from its nearest neighbor within a specified radius.

2. Core Points:
Points: Points with a sufficient number of neighbors within the specified radius are considered core points.

3. Borders Points:
Points that are reachable from a core point but do not have enough neighbors to be considered core.

4. Optics Ordering:
algorithm computes an ordering of the points based on their reachability distances. This ordering is used to determine the clusters later.

5. Clustering:
Finally, clusters are formed based on the reachability distances and a minimum number of points required to form a cluster.

```python
from sklearn.cluster import OPTICS
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
X = np.random.rand(100, 2)
```

```python
X[0:5]
```

```python
# Fit the OPTICS clustering algorithm
model = OPTICS(min_samples=5, xi=0.05, min_cluster_size=0.1)
model.fit(X)

# Plotting the clusters
colors = ['blue', 'red', 'green', 'yellow', 'purple', 'orange']
plt.figure(figsize=(8, 6))
for i, color in enumerate(colors):
    mask = model.labels_ == i
    plt.scatter(X[mask, 0], X[mask, 1], c=color, label=f'Cluster {i}')

plt.legend()
plt.title('OPTICS Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Example 2:

```python
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# Generate synthetic data
X, y= make_moons(n_samples=300, noise=0.05, random_state=42)

# Spectral Clustering
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=42)
labels = model.fit_predict(X)

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Spectral Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

###  Example 3

```python
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering
```

```python
# Generate synthetic data with two moons
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Create a spectral clustering object
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=42)

# Fit and predict clusters
clusters = model.fit_predict(X)
```

```python
clusters[0:5]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', s=50)
plt.title("Spectral Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
