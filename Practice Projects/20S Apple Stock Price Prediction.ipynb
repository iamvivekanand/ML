```python
import plotly.graph_objects as go
```

```python
data = pd.read_csv("AAPL.csv")
data.head()
```

```python
df=data.copy()
df.columns=['date','open','high','low','close','adj close','volume']
```

```python
dataa=[go.Candlestick(x=df['date'],
                     open=df['open'],
                     low=df['low'],
                     high=df['high'],
                     close=df['close'])]
fig = go.Figure(dataa)
fig.update_layout(title='Apple Price Prediction',xaxis_rangeslider_visible=False)
fig.show()
```

```python
df2=df.query('close>=176.210007').sort_values(by='close',ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df2['date'],y=df2['close'],color='green')
plt.title('Apple Top 5 Closing Stock with Date',fontdict={'color':'black','size':20,'family':'times new roman'})
plt.xlabel('DATE',fontdict={'color':'blue','size':15,'family':'times new roman'})
plt.ylabel('CLOSE',fontdict={'size':15,'color':'blue','family':'times new roman'})
plt.xticks(rotation=90)
plt.show()
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(df, date_col='date', value_col='close', id_col=None)
pred = model.predict()
forecast = pred.forecast
print(forecast)
```
