Normal distribution:
- Normal distribution is the most common type of distribution in statistical analyses,
- It has two parameters 'Mean' and 'Standard Deviation'
- Mean is the central tendency of the distribution, U can calculate by summin up all values then deviding it by total no of observation.
- The standard deviation is a measure of variability. It defines the width of the normal distribution.
- Standard Deviation is often written as 'std' in short,it tells how far the values of observation can fall on either side of the mean value.

## Feature scaling
- Feature Scaling is a technique to standardize the independent features present in the data in a fixed range.

EUCLIDIAN DISTANCE:
- square root of the sum of squares of differences between the coordinates (feature values – Age, Salary, BHK Apartment) of data point and centroid of each class. 
- let there be two points (x1,y1) and (x2,y2)

 √[(x2 – x1)2 + (y2 – y1)2].

MANHATTAN DISTANCE:
- sum of absolute differences between the coordinates (feature values) of data point and centroid of each class.

The Manhattan Distance between two points (X1, Y1) and (X2, Y2) is given by |X1 – X2| + |Y1 – Y2|.

#### Need of feature scaling
Feature scaling is essential for machine learning algorithms that calculate distances between data. If not scale, the feature with a higher value range starts dominating when calculating distances

Feature scaling is performed during data pre-processing to handle highly varying magnitudes or values or units

Without Feature scaling an algorithm can assume 3000 meters > 3KM

```python
# Robust Scaler:
```

- uses the interquartile range, rather than the min-max, so that it is robust to outliers
- removes the median and scales the data according to the quantile range
- The IQR is the range between the 1st quartile (25th quantile) and the 3rd quartile (75th quantile).
- has no predetermined range, unlike Min-Max Scaler
