```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
# avoiding warnings display
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("/kaggle/input/insurence-dataset-latest/insurance.xlsx")
df=data.copy()
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
ss=StandardScaler()
le=LabelEncoder()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
x=df.drop('expenses',axis=1)
y=df.expenses
```

```python
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x, y, test_size = 0.25, random_state=41)
```

```python
# SVM
from sklearn.svm import SVR
svm=SVR()
svm.fit(xtrain,ytrain)
print('Train Accuracy: ',svm.score(xtrain,ytrain))
print('Test Accuracy: ',svm.score(xtest,ytest))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=500,)
rf.fit(xtrain,ytrain)
print('Train Accuracy: ',rf.score(xtrain,ytrain))
print('Test Accuracy: ',rf.score(xtest,ytest))
```

```python
# AdaBoost
from sklearn.ensemble import AdaBoostRegressor
adb=AdaBoostRegressor(base_estimator=rf,n_estimators=300,)
adb.fit(xtrain,ytrain)
print('Train Accuracy: ',adb.score(xtrain,ytrain))
print('Test Accuracy: ',adb.score(xtest,ytest))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(xtrain,ytrain)
print('Train Accuracy: ',dt.score(xtrain,ytrain))
print('Test Accuracy: ',dt.score(xtest,ytest))
```

>Out of the above 4 models, Random Forest has got the highest training and testing accuracy combination. Although, Decision Tree gave highest training accuracy of 99% but it failed to give better testing accuracy, it has been overfitted. So We will use Random Forest for Model Prediction
